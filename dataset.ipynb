```python
def split_train_test(data):
    fid = open(data,"r")
    li = fid.readlines()
    fid.close()
    random.shuffle(li)
    train = li[0:int(len(li) * 0.6)]
    val = li[int(len(li) * 0.6): int(len(li) * 0.8)]
    tes = li[int(len(li) * 0.8): len(li)]
    return train, val, tes
```

```python
def write_to_csv(path, category):
    fid = open(path,"w")
    fid.writelines(category)
    fid.close()
```

```python
train, val, tes = split_train_test('./data.csv')
write_to_csv('train.csv', train)
write_to_csv('val.csv', val)
write_to_csv('tes.csv', tes)
```
